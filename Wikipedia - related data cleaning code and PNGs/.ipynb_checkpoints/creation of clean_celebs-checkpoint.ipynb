```python
 %matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
glo_data = "Global.csv"
citi_data = "USA.csv"
```

```python
global1 = pd.read_csv(glo_data)
usa = pd.read_csv(citi_data)
#usa.head()
#making working dataframe to play with
usa_work = usa
usa_work.head()
```

```python
#splitting fname and lname on the basis of space between names
usa1 = usa_work["name"].str.split(" ", n = 1, expand = True) 
usa1.head()

# making separate first name column from new data frame 
usa_work["firstName"]= usa1[0]
usa_work["lastName"]= usa1[1]
#dropping the name column after it was replaced by 
usa_work.drop(columns =["name"], inplace = True) 
  
usa_work.head()
```

```python
usa_work1 = usa_work.drop(columns=['ID','birthstate','countryName','countryCode', 'gender','occupation'])
```

```python
#usa_work1.head()
usa_work2 = usa_work1[usa_work1['birthyear'] > 1909]
usa_work2.head()
```

```python
#create bins for decades
#Create the bins in which Data will be held
#Bins are 0, 59, 69, 79, 89, 100.

bins = [1909 ,1919, 1929, 1939, 1949, 1959, 1969, 1979, 1989, 1999, 2009, 2019]
# Create the names for the four bins
group_names = ["10s", "20s", "30s", "40s", "50s", "60s", "70s","80s","90s","00s", "2010s"]
usa_work2["decade"] = pd.cut(usa_work2["birthyear"], bins, labels=group_names)
usa_work2.drop(columns =["birthyear"], inplace = True)

```

```python
usa_work3 = usa_work2
usa_work3.head()
```
