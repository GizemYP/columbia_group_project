```python
import pandas as pd
```

```python
#read one file 
df = pd.read_csv('SSN_Data/AK.TXT', names=['State', 'Gender', 'Year', 'Name', 'Count'])
df.head()
```

```python
#I created a different dataframe with subset of the "df" dataframe
baby_name_F_70 = df[(df.Year >= 1970) & (df.Year <= 1979) & (df.Gender =="F")]
```

```python
#this new dataframe is helping me figure out if the count for the below dataframe is correct. By checking the number of
#counts for each name within the years of 1970 - 1979
#specific_name = df[(df.Year >= 1970) & (df.Year <= 1979) & (df.Name =="Jennifer")]
#specific_name
```

```python
#grouped by "Name", "State and Gender"
count_name = baby_name_F_70.groupby(["Name","State", "Gender"])
count_name
```

```python
#Sum of the amount of times a name was chosen in the state of AK
total_count_of_name = count_name.sum()["Count"]
#total_count_of_name
```

```python
#Created a new dataframe with the group by of "Name" and "State" and aggregated by SUM. I also added a sort
#by ["Count"] and ordered it by descending
total_count_of_names_df = pd.DataFrame(total_count_of_name).sort_values(by=["Count"], ascending=False).rename(columns={"Name":"Name", "State": "State", "Count":"Total"})
                                                                                                        
total_count_of_names_df.head(10)
```

```python
#I'm doing the same steps for the Male names 
baby_name_M_70 = df[(df.Year >= 1970) & (df.Year <= 1979) & (df.Gender =="M")]
#baby_name_M_70.head(10)
```

```python
#baby_name_F_70 = df[(df.Year >= 1970) & (df.Year <= 1979) & (df.Gender =="M")]
#baby_name_F_70.head()
```

```python
count_name_m = baby_name_M_70.groupby(["Name", "State", "Gender"])

total_count_name_m = count_name_m.sum()["Count"]
total_count_name_m
```

```python
total_count_of_names_m_df = pd.DataFrame(total_count_name_m).sort_values(by=["Count"], ascending=False).rename(columns={"Name":"Name", "State":"State", "Count":"Total"})
                                                                                                        
total_count_of_names_m_df.head(10)
```

```python
#This dataframe is combining the top 10 of male and female names.
Male_female_name_count_1970s = pd.concat([total_count_of_names_df.head(10), total_count_of_names_m_df.head(10)])
Male_female_name_count_1970s
```
