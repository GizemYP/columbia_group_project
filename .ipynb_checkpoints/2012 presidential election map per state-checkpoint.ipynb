```python
#Call on dependencies 
import pandas as pd 
import folium 
import os
import vincent
```

```python
# This variable will represent the states data with all the coordinates
states = os.path.join('Data','us_states.json')
```

```python
#This variable will represent the 2012 presidential political party results per state
political_party_data = os.path.join("Data","2012 presidential election per state.xlsx")
```

```python
#This variable will represent our election data going through the pandas read_csv function
state_data = pd.read_excel(political_party_data)
state_data.head()
```

```python
#Show the class 
head = state_data.head()
html = head.to_html(classes='table table-striped table-hover table-condensed table-responsive')
popup = folium.Popup(html)
```
