```python
import pandas as pd
import glob

#create a dataframe
name_df = pd.DataFrame()

#open, read all text files and name the column labels
for files in glob.glob('../SSN_data/*.txt'):
    comb_data = pd.read_csv(files,names=['state', 'gender', 'year', 'name', 'count'])
#concatenate text files in the dataframe
    name_df = pd.concat([name_df,comb_data],axis=0,ignore_index=True)
name_df.head()
```

```python
#convert year in df to be a string
name_df['year']=name_df['year'].astype(str)

#valid choices for election year input
choices=list(["1972", "1976", "1980", "1984", "1988", "1992", "1996", "2000", "2004",
              "2008", "2012", "2016"])
while True:
    # ask user to select an election year
    year=str(input("Please choose election year between 1972 and 2016: "))
    print("You entered:", year)
    
    if year in choices:
      #create a selected election year DF by filtering total DF on selected year
        selectedDF=name_df.loc[(name_df['year']==year)].reset_index(drop=True)
        print(selectedDF.head())
        break
    else:
        print("Not an election year.  Should I google it for you? Try again... ")
```

```python
#10 most pupulart names - for each gender, sorted from the most popular in decending order
top10names=selectedDF.groupby(['gender','name'])['count'].sum().reset_index().sort_values(by='count',ascending=False).reset_index(drop=True)
top10names_sorted=top10names.groupby('gender').head(10)
top10names_sorted.sort_values(['gender','count'], ascending=False).reset_index(drop=True) 
```
